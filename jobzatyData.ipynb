```python
import csv 
import requests
from bs4 import BeautifulSoup
from itertools import zip_longest
```

```python
job_titles=[]
companys_names=[]
citys=[]
links=[]
# scrape data 93 home page of the website
for i in range(1,94):
    result= requests.get(f"https://www.jobzaty.com/jobs?page={i}")
    src=result.content
    soup=BeautifulSoup(src,'lxml')
    # Extract the job ad title
    job_title=soup.find_all('h2')
    # Extract the company name
    company=soup.find_all('div',class_='company')
    # Extract the city
    city=soup.find_all('span',class_='newCity')
    # Extract the ad link
    linkh=soup.find_all('h2')
    for l in linkh:
        link=l.find('a',href=True)
        links.append(link['href'])
    #append job titles,companys, and citys in lists
    for n in range(len(company)):
        job_titles.append(job_title[n].text.strip())
        companys_names.append(company[n].text.strip())
        citys.append(city[n].text.strip())
```

```python
print(len(job_titles))
print(len(companys_names))
print(len(citys))
print(len(links))
```

```python
career_levels=[]
experiences=[]
dates=[]
# scrape data from each ad page
for i in range(len(links)):
    result= requests.get(links[i])
    src=result.content
    soup=BeautifulSoup(src,'lxml')
    # First extract the list of the job details 
    jobdetailes=soup.find('ul',class_='jbdetail').find_all('li')
    # Extract the career level from the list
    career_level=jobdetailes[3].find('span').text.strip()
    # Extract the experience from the list
    experience=jobdetailes[5].find('span').text.strip()
    # Extract the date from the list
    date=jobdetailes[8].find('span').text.strip()
    #append career levels,experiences, and dates in lists
    career_levels.append(career_level)
    experiences.append(experience)
    dates.append(date)
```

```python
print(career_levels,experiences,dates)
```

```python
print(len(career_levels))
print(len(experiences))
print(len(dates))
```

```python
#write data to csv file
file_lists=[job_titles,companys_names,citys,career_levels,experiences,dates,links]
exported=zip_longest(*file_lists)
with open('jobzatyData.csv','w',newline='') as file:
    wr=csv.writer(file)
    wr.writerow(['الوظيفة','اسم الشركة','المدينة','المستوى الوظيفي','الخبرة','التاريخ','الرابط'])
    wr.writerows(exported)
```
